```python
from numpy import random, cosh

from matplotlib import pyplot as plt

N = 7000

a = 0
b = 10
b1 = 5

x = random.uniform(a, b, N)
y = random.uniform(a, b1, N)

plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.title('$cosh(x/2)$ integralis')

plt.plot(x, y, 'ko')

N1 = 0

for i in range(N):
    if y[i] < cosh(x[i] / 2):
        plt.plot(x[i], y[i], 'r')
        N1 = N1 + 1
    else:
        plt.plot(x[i], y[i], 'd')
        
plt.show()

S = (b - a) * (b1 - a)
newS = float(S * N1 / N)
print("Integracijas rezultaats:", newS)
```
